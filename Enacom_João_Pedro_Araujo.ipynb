```python
from sympy import *
import matplotlib as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
x1 = np.linspace(-10, 10, num=100)
x2 = np.linspace(-10, 10, num=100)
x1, x2 = np.meshgrid(x1, x2)

def func(x1, x2):
  return pow(x1,2) - (2*x1) + 1 + (2*pow(x2,2))

z = pow(x1,2) - (2*x1) + 1 + (2*pow(x2,2))

pt_otimo = [1, 0]
```

```python
fig, ax = plt.subplots()
CS = ax.contour(x1, x2, z,10,colors='k')
ax.clabel(CS, inline=0.5, fontsize=8)
ax.plot(pt_otimo[0], pt_otimo[1], func(pt_otimo[0], pt_otimo[1]), 'x', c='r')
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X1, X2, Z = axes3d.get_test_data(0.01)
ax.plot_surface(x1, x2, z, cmap=cm.nipy_spectral)
```
